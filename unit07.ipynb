```python
import csv
f = open('age.csv')
data = csv.reader(f)
result = []


for row in data :
    if '신도림' in row[0] :
        for i in row[3:] :   #row 옆에 index값 : 를 붙여주면 해당 인덱스 이후 나머지 까지가 나옴 
            result.append(int(i))
print(result)
```

```python
import csv
f = open('age.csv')
data = csv.reader(f)
result = []


for row in data :
    if '신도림' in row[0] :
        for i in row[2:] :   #row 옆에 index값 : 를 붙여주면 해당 인덱스 이후 나머지 까지가 나옴 
            result.append(int(i))
print(result)

import matplotlib.pyplot as plt
plt.style.use('ggplot')
plt.plot(result)
plt.show()
```

```python
import csv

f = open('age.csv')
data = csv.reader(f)
result = []
name = input('인구 구조가 알고 싶은 지역의 이름(읍면동 단위)를 입력해라 : ')  #input으로 입력받기 

for row in data :
    if name in row[0] :       #포함된 값 찾기
        for i in row[2:] :
            result.append(int(i.replace(',','')))
            
import matplotlib.pyplot as plt
plt.style.use('ggplot')  #그래프 적용
plt.rc('font',family='Malgun Gothic')   
plt.title(name + ' 지역의 인구 구조')
plt.plot(result)
plt.show()
```
