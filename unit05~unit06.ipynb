```python
import csv
import matplotlib.pyplot as plt
f = open('seoul.csv')
data = csv.reader(f)
next(data) #next 함수를 사용해 헤더부분 제외 
result = [] # 최고 기온 데이터를 저장할 리스트 생성


for row in data :
    if row[-1] != '' :    #기온값 존재시 
        if row[0].split('-')[1] == '08' :   #8월에 해당하는 값이면
            result.append(float(row[-1]))   #result 리스트에 기온 값 추가
#plt.figure(figsize = (10,2))
plt.plot(result, 'hotpink')
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt
f = open('seoul.csv')
data = csv.reader(f)
next(data) #next 함수를 사용해 헤더부분 제외 
result = [] # 최고 기온 데이터를 저장할 리스트 생성

for row in data :
    if row[-1] != '' :
        if row[0].split('-')[1] == '02' and row[0].split('-')[2] == '14' :   #2월 14일 
            result.append(float(row[-1]))
plt.plot(result, 'hotpink')
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv')
data = csv.reader(f)
next(data)
high = []   #최고 기온 값 저장
low = []   #최저 기온 값 저장 

for row in data :
    if row[-1] != '' and row[-2] != '' :
        if 1983 <= int(row[0].split('-')[0]) :  #1983년 이후 데이터면
            if row[0].split('-')[1] == '02' and row[0].split('-')[2] == '14' : #2월 14일이면
                high.append(float(row[-1]))
                low.append(float(row[-2]))

plt.rc('font', family = 'Malgun Gothic')  #show 문장 앞에 써주어야함
plt.title('2월 14일의 기온 변화 그래프')  #show 문장 앞에 써주어야함
plt.plot(high, 'hotpink' , label = 'high')
plt.plot(low, 'skyblue' , label = 'low')
plt.legend()   #범례표시 plot에 label로 준 값이 나옴
plt.show()

                
```

```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv')
data = csv.reader(f)
next(data)

aug = []
jan = []

for row in data :
    month = row[0].split('-')[1]
    if row[-1] != '' :
        if month == '08' :
            aug.append(float(row[-1]))
        if month == '01' :
            jan.append(float(row[-1]))
        
plt.hist(aug, bins=100, color = 'r', label='Aug') 
plt.hist(jan, bins=100, color = 'b', label='Jan') 
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import csv

f = open('seoul.csv')
data = csv.reader(f)
next(data)

day = []   #일별데이터 저장할 day 생성
for i in range(31) :
    day.append([])     #day 리스트 내 31개 리스트 생성
    
    
for row in data :
    if row[-1] != '' :
        if row[0].split('-')[1] == '08' :    #8월 일때 
            day[int(row[0].split('-')[2])-1].append(float(row[-1]))
            
plt.style.use('ggplot') #그래프 스타일 지정
plt.figure(figsize = (10,5), dpi = 300) #그래프 크기 수정
plt.boxplot(day, showfliers=False) #아웃라이어 값 생략  = 이상치 값이 보이지 않게 설정 

plt.show() 
```
