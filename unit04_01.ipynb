```python
import matplotlib.pyplot as plt #matplotlib 홈페이지의 코드에서도 plt라는 별명 사용
plt.plot([10,20,30,40]) #입력한 리스트의 값은 y축의 값으로 입력됨 (한축만 등록 시)
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.title('legend') 
plt.plot([10,20,30,40], label ='asc') #증가를 의미하는 asc범례
plt.plot([40,30,20,10], label ='desc') #감소를 의미하는 asc범례
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.title('legend') 
plt.plot([10,20,30,40], label ='asc') #증가를 의미하는 asc범례
plt.plot([40,30,20,10], label ='desc') #감소를 의미하는 asc범례
plt.legend(loc = 5) #범례 위치바꾸기
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.title('color') 
plt.plot([10,20,30,40], color='skyblue', label ='skyblue') 
plt.plot([40,30,20,10],'pink', label ='pink')
plt.legend() 
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.title('linestyle') 
plt.plot([10,20,30,40], color='r', linestyle='--', label ='edashed')   #빨간색 dashed 그래프
plt.plot([40,30,20,10], color='g', ls=':', label ='dotted')
plt.legend() 
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.title('marker') 
plt.plot([10,20,30,40], 'r',  label ='circle')   #빨간색 원형 마커 그래프
plt.plot([40,30,20,10], 'g^', label ='triangle up') #초록색 삼각형 마커 그래프
plt.legend() 
plt.show()
```
