```python
import csv

f = open('age1.csv')
data = csv.reader(f)
result = []
name = input('인구 구조가 알고 싶은 지역의 이름(읍면동 단위)를 입력해라 : ')  #input으로 입력받기 

for row in data :
    if name in row[0] :       #포함된 값 찾기
        for i in row[3:] :
            result.append(int(i.replace(',','')))
            
import matplotlib.pyplot as plt
plt.style.use('ggplot')  #그래프 적용
plt.rc('font',family='Malgun Gothic')   
plt.title(name + ' 지역의 인구 구조')
plt.plot(result)
plt.show()
```

```python
import csv
f = open('gender.csv')
data = csv.reader(f)
m = []
f = []

for row in data :
    if '신도림' in row[0] :
        for i in row[3:104] :
            m.append(-int(i))   #한 쪽 값을 음수로 바꾸어 보기 편하게 그래프 형식 변경
        for i in row[106:] :
            f.append(int(i))
            
import matplotlib.pyplot as plt
plt.rc('font',family = 'Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False  #꺠진 마이너스 수정 
plt.title('신도림 지역의 남녀 성별 인구 분포')
plt.barh(range(101),m,label = '남성')
plt.barh(range(101),f,label = '여성')
plt.legend
plt.show()
```

```python
import csv
f = open('gender.csv')
data = csv.reader(f)
m = []
f = []
name = input('찾고 싶은 지역의 이름을 알려주세요 : ')


for row in data :
    if name in row[0] :
        for i in row[3:104] :
            m.append(-int(i.replace(',','')))   #i의 값안에 콤마가 있어 int로 인식 할 수 없어 replace 작업 
        for i in row[106:] :
            f.append(int(i.replace(',','')))   #i의 값안에 콤마가 있어 int로 인식 할 수 없어 replace 작업 
        break
            
import matplotlib.pyplot as plt
plt.style.use('ggplot')
plt.figure(figsize=(10,5), dpi=300)
plt.rc('font',family = 'Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False  #꺠진 마이너스 수정 
plt.title(name + ' 지역의 남녀 성별 인구 분포')
plt.barh(range(101),m,label = '남성')
plt.barh(range(101),f,label = '여성')
plt.legend
plt.show()
```

```python
import csv

f = open('gender.csv')
data = csv.reader(f)
size = []
name = input('찾고 싶은 지역의 이름을 알려주세요 :')

for row in data :
    if name in row[0] :
        m = 0 #변수 초기와
        f = 0 #변수 초기와
        for i in range(101) :
            m += int(row[i+3].replace(',',''))  #누적 더하기
            f += int(row[i+106].replace(',',''))
        break 
size.append(m)  #size 리스트에 더하기
size.append(f)
print(size)

import matplotlib.pyplot as plt
plt.rc('font', family = 'Malgun Gothic')
color = ['crimson','darkcyan']
plt.axis('equal')
plt.pie(size, labels=['남','여'], autopct ='%.1f%%', colors=color, startangle=90)
plt.title(name + '지역의 남녀 성별 비율')
plt.show()
```

```python
import csv
import math

f = open('gender.csv')
data = csv.reader(f)
m = []
f = []
size = []
name = input('궁금한 동네를 입력해주세요 :')
for row in data :
    if name in row[0] :
        for i in range(3,104) :
            m.append(int(row[i].replace(',','')))
            f.append(int(row[i+103].replace(',','')))
            size.append(math.sqrt(int(row[i].replace(',',''))+int(row[i+103].replace(',',''))))
        break
        
import matplotlib.pyplot as plt
plt.style.use('ggplot')
plt.rc('font',family='Malgun Gothic')
plt.figure(figsize=(10,5),dpi=300)
plt.title(name+' 지역의 성별 인구 그래프')
plt.scatter(m,f,c=range(101), alpha=0.5, cmap='jet')
plt.colorbar()
plt.plot(range(max(m)),range(max(m)), 'g') #추세선 추가 g = green
plt.xlabel('남성 인구수')
plt.ylabel('여성 인구수')
plt.show()
```
